# Script to run Pangolin from Jupyter Notebook 

To run Notebook: 

1) Have the notebook in the same directory as the directories: 'Sequence' and 'Final_Reports'<br>
2) Inside the 'Sequence' directory, place the .fasta file that pangolin will use fo analysis. <br>
3) Both "Final_Report" and "Shell" Directories must be empty before you start running this notebook. <br>
4) After the script has ben ran the shell file created will me moved into the 'Shell' directory. <br>

# 1) Retrieve FASTA file name

This snippet of code gets the file name from the FASTA file in the Sequence directory. 

```python
## Function Retrieves Fasta file name. 
## This file name will be used in the creation of 
## a bash script.
#def fasta_file_name():
from pathlib import Path
path=Path("Sequence/")
file_name=''
for entry in path.iterdir():
    #Deals with .DS_Store file automatically made in MacOS
    if entry.name.startswith("."):
        continue
    #print("*"*100)
    else:
        print("Using FASTA file:\n" + entry.name+ "\n" )
        file_name=entry.name
        break

        
        
```

# 2) Write shell script that will run pangolin

```python
#shell_name=file_name[0:len(file_name)-6]

shell_name="pangolin_shell.sh"

new_file=open(shell_name,'w+')
new_file.write("#!/bin/sh\n\n")
new_file.write("cd ./Sequence/\n")
new_file.write('eval "$(conda shell.bash hook)"\n')
new_file.write("conda activate pangolin\n")
new_file.write("pangolin " + file_name+'\n')
new_file.close()

print("Shell Script Looks Like:\n")
! cat pangolin_shell.sh
```

### 2.1) Make Pango shell script executable 

```python
! chmod 700 pangolin_shell.sh
```

### 2.2) Run Pagnolin from Conda Envirornment
sournce: https://stackoverflow.com/questions/34534513/calling-conda-source-activate-from-bash-script

```python
### Run pangolin_shell.sh script.
import subprocess
import time

#Start Counting time elapsed. 
start = time.time()

subprocess.call(["./pangolin_shell.sh"])


#Record en of elapsed time. 
end = time.time()

#Compute seconds elapsed. 
elapsed= end-start
print("Seconds elapsed to comlpete Pangolin\n" + str(elapsed) + " Seconds ")
```

### 2.3) Move Pango lineage report to Final_Report Directory

Note: For this part not to fail, the Final_Report directory must not already contain a file with the same name as the one being moved into the directory. 

In this part, the Pango linage report is also renamed to the file name: pango_lineage_report.csv

```python
import os
import shutil
#get working directory 
pwd=os.getcwd()

#move file from source diectory

source=pwd+"/Sequence/lineage_report.csv"

#Destitation directory 
destinaiton= pwd+"/Final_Report/pango_lineage_report.csv"

#Move lineage_report.csv to Final Report Directory
dest=shutil.move(source, destinaiton)

print("lineage_report.csv has been renamed to'pango_lineage_report.csv'\nAnd moved to the directory: /Final_Report")
```

### 2.4) Move Pago Shell script to 'Shell' Directory

```python
#move file from source diectory
source=pwd+'/'+shell_name


#Destitation directory 
destinaiton= pwd+"/Shell/"

#Move lineage_report.csv to Final Report Directory
dest=shutil.move(source, destinaiton)

print(shell_name+" has been moved to the directory\n'Shell'")
```

# 3)  Write shell script that will run UShER pangolin

```python
#shell_name=file_name[0:len(file_name)-6]

shell_name="usher_pangolin_shell.sh"

new_file=open(shell_name,'w+')
new_file.write("#!/bin/sh\n\n")
new_file.write("cd ./Sequence/\n")
new_file.write('eval "$(conda shell.bash hook)"\n')
new_file.write("conda activate pangolin\n")
new_file.write("pangolin --usher " + file_name+'\n')
new_file.close()

print("Shell Script Looks Like:\n")
! cat usher_pangolin_shell.sh
```

### 3.1) Make Pango shell script executable 

```python
! chmod 700 usher_pangolin_shell.sh
```

### 3.2) Run UShER Pagnolin from Conda Envirornment

Note: Using UShER tag took more than one hour to run 485 Sequences. 

```python
### Run pangolin_shell.sh script.
import subprocess
import time

#Start Counting time elapsed. 
start = time.time()

#Start Pangolin UShER timer. 
subprocess.call(["./usher_pangolin_shell.sh"])

#Record en of elapsed time. 
end = time.time()

#Compute seconds elapsed. 
elapsed= end-start
print("Seconds elapsed to comlpete Usher\n" + str(elapsed) + " Seconds ")
```
